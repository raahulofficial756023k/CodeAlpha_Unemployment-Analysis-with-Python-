```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv(r'D:\CodeAlpha\TASK 2\Unemployment in India.csv')

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Average unemployment rate by region
unemployment = df.groupby('Region')["Estimated Unemployment Rate (%)"].mean()

# Plot
plt.figure(figsize = (10,5))
unemployment.sort_values(ascending=False).plot(kind="bar")
plt.title("Average Unemployment Rate by Region")
plt.ylabel("Unemployment Rate (%)")
plt.xlabel("Region")
plt.show()
```

```python
# Data Cleaning
df.columns = df.columns.str.strip()
df["Data"] = pd.to_datetime(df["Date"], dayfirst = True)

# Check missing values
print(df.isnull().sum())

# Data Exploration 
print(df.describe())

# Monthly Unemployment trend
trend = df.groupby("Date")["Estimated Unemployment Rate (%)"].mean()

# Visualization
plt.figure(figsize= (20,5))
plt.plot(trend.index, trend.values, marker='o')
plt.title("Umemployment Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

```python

# Create covid_data
covid_data = df.groupby("Date")["Estimated Unemployment Rate (%)"].mean()
```

```python
#Plot
plt.figure(figsize = (20,5))
plt.plot(covid_data.index, covid_data.values, marker='o')

# Covid lockdown period
plt.axvline(pd.to_datetime("2020-03-24"), 
            color = 'red',
            linestyle = '--',
            label = 'COVID Lockdown')
plt.title("Impact of COVID on Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")
plt.legend()
plt.grid(True)

plt.show()
```

Identify key patterns or seasonal trends in the data. 

```python
# Extract month name
df["Month"] = df["Date"].dt.month_name()

# Average Unemployment by month
monthly_trend = df.groupby("Month")["Estimated Unemployment Rate (%)"].mean()

# Arrange Months in calender order
month_order = ["January", "February", "March", "April", 'May',
               "June", "July", "August", "September", "October",
               "November", "December"]
monthly_trend = monthly_trend.reindex(month_order)

# Plot
plt.figure(figsize = (20,5))
plt.plot(monthly_trend.index, monthly_trend.values, marker = "o")
plt.title("Average Unemployment Rate by Month")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```
